```python
# prompt: read xml file and parse it, consider root =publicExecutionReport and parse whole xml

import xml.etree.ElementTree as ET
import pandas as pd

# Assuming the XML file is named 'data.xml'
tree = ET.parse('/content/drive/My Drive/Colab Notebooks/anomaly/ird-ex01-vanilla-swap.xml')
root = tree.getroot()

# Extract data from the root element and all its descendants
data = []
row = {}

def extract_data(element, parent_tag=""):
  for i, child in enumerate(element):
    tag = f"{parent_tag}.{child.tag}_{i}" if parent_tag else f"{child.tag}"
    if len(child) == 0:
      row[tag] = child.text
    else:
      extract_data(child, tag)

extract_data(root)
data.append(row)

# Create DataFrame
df = pd.DataFrame(data)
df
```
