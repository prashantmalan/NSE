```python
import pandas as pd
import yfinance as yf

# Create an empty DataFrame
df_flattened = pd.DataFrame()
df_flattened1 = pd.DataFrame()
for symbol in df_nse['SYMBOL']:
    ticker = yf.Ticker(symbol + '.NS')
    # Iterate over each key-value pair in ticker.info
    for key, value in ticker.info.items():
        print(key)
        print (value)
        # Check if the value is a dictionary
        if isinstance(value, dict):
            # Flatten the nested dictionary into a DataFrame
            df_nested = pd.json_normalize(value)
            # Add a prefix to the column names to indicate the key
            df_nested = df_nested.add_prefix(key + '_')
            # Append the flattened DataFrame to the main DataFrame
            df_flattened = pd.concat([df_flattened, df_nested], axis=1)
            df_flattened1=pd.concat([df_flattened1, df_flattened], axis=0, ignore_index=True)
            
        else:
            # If the value is not a dictionary, create a single-row DataFrame
            df_single = pd.DataFrame({key: [value]})
            # Append the single-row DataFrame to the main DataFrame
            df_flattened = pd.concat([df_flattened, df_single], axis=1)
            df_flattened1=pd.concat([df_flattened1, df_flattened], axis=0,ignore_index=True)
            #df_flattened.set_index('symbol')
    # Print the flattened DataFrame
#df_flattened=df_flattened.set_index('symbol')
#df_flattened.to_csv('fundamental.csv')
print(df_flattened)
df_flattened.to_csv('fundamental.csv')
```
